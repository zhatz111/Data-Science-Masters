### Estimate the Regression Coefficients using matrix form

```python
X_reg = q2_table[["x1", "x4"]].values
y = q2_table["y"].values

X = np.column_stack((np.ones(X_reg.shape[0]), X_reg))  # Add intercept term

# Degrees of freedom
dfn = X.shape[1] - 1  # Degrees of freedom for the numerator (number of predictors)
dfd = len(y) - X.shape[1]  # Degrees of freedom for the denominator (n - k - 1)

beta = np.linalg.inv(X.T @ X) @ X.T @ y
print("Estimated coefficients (beta):", beta)
```

### Calculate the sum of squares and residual variance for the model

```python
y_bar = np.mean(y)
y_pred = X @ beta
sst = np.sum((y - y_bar) ** 2)
ssr = np.sum((y_pred - y_bar) ** 2)
sse = np.sum((y - y_pred) ** 2)
res_variance = sse / (len(y) - X.shape[1])
print("Total Sum of Squares (SST):", sst)
print("Regression Sum of Squares (SSR):", ssr)
print("Error Sum of Squares (SSE):", sse)
print("Residual Variance (s^2):", res_variance)
```

### Calculate the Covariance Matrix, Standard Error Vector, t-statistics, and p-values for each Regressor

```python
variance = res_variance * np.linalg.inv(X.T @ X)
std_error = np.sqrt(np.diag(variance))
print("Variance-Covariance Matrix of coefficients:\n", variance)
print()
print("Standard Errors of coefficients:", std_error)
```

```python
t_stats = beta / std_error
print("t-statistics for each coefficient:", t_stats)
```

```python
# Use the t-distribution to find the two-tailed p-values
p_value_t = [t.sf(abs(ti), df=dfd) * 2 for ti in t_stats]
print("p-values for each coefficient:", p_value_t)
```

### Calculate the F-statistic and p-value from the f distribution

```python
f_statistic = ((ssr/dfn) / res_variance)
# Use the F-distribution to find the p-value
p_value_f = f.sf(f_statistic, dfn, dfd)
print("F-test statistic:", f_statistic)
print("p-value:", p_value_f)
```

### Correlation between regressors

```python
cov_beta = res_variance * np.linalg.inv(X.T @ X)
corr = cov_beta[1,2] / np.sqrt(cov_beta[1,1]*cov_beta[2,2])

print("Covariance matrix between regressors:\n", cov_beta)
print()
print("Correlation between regressors x1 and x4:", corr)
```
