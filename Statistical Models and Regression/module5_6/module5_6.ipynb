### Question 3

```python
filename2 = "Module 5-6 Assignment Problem 3 Data.txt"
df_q3 = pd.read_csv(Path(path, filename2), sep='\s+')
df_q3.head()
```

```python
# randomly select one sample from each combination of A and B
from itertools import product

q3_samp = df_q3.iloc[:6, :].copy()
for idx, (a, b) in enumerate(product(df_q3["A"].unique(), df_q3["B"].unique())):
    mask = (df_q3["A"] == a) & (df_q3["B"] == b)
    df_sample = df_q3.loc[mask, :].sample(n=1, random_state=SEED)
    q3_samp.iloc[idx, :] = df_sample.iloc[0].values

q3_samp
```
