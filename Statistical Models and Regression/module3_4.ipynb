## Question 3

### Part A

```python
import numpy as np

# --- Define a multiple linear regression model function ---
def ols_model(x_6, x_7, beta_0, beta_6, beta_7):
    """
    Computes predicted y values for a multiple linear regression model:
        y_hat = beta_0 + beta_6 * x_6 + beta_7 * x_7
    Parameters:
        x_6     : first predictor (total solvent)
        x_7     : second predictor (hydrogen consumption)
        beta_0  : intercept
        beta_6  : coefficient for x_6
        beta_7  : coefficient for x_7
    Returns:
        predicted y values
    """
    return beta_0 + beta_6*x_6 + beta_7*x_7

# --- Select predictors and response from the randomized subset of data ---
X = rand_df[["x6", "x7"]].values  # predictor columns
y = rand_df["y"].values           # response variable

# --- Add a column of ones to X to account for the intercept term ---
X = np.hstack((np.ones((X.shape[0], 1)), X))  # shape becomes (n, 3)

# --- Estimate regression coefficients using the normal equation ---
# beta_hat = (X^T X)^(-1) X^T y
beta_hat = np.linalg.inv(X.T @ X) @ X.T @ y

# --- Calculate predicted values using the estimated coefficients ---
y_preds = ols_model(
    X[:,1],       # x_6 values
    X[:,2],       # x_7 values
    beta_hat[0],  # intercept
    beta_hat[1],  # beta_6
    beta_hat[2]   # beta_7
)

print(f"Intercept: {beta_hat[0]:0.04f}")
print(f"Beta 6: {beta_hat[1]:0.04f}")
print(f"Beta 7: {beta_hat[2]:0.04f}")
```

### Part B

```python
# define the sample size and number of predictors
n = 20
p = 2

# --- Total sum of squares decomposition ---

# Total Sum of Squares (SST): total variation in y about its mean
SST = np.sum((y - np.mean(y))**2)

# Regression Sum of Squares (SSR): variation in fitted values about the mean of y
SSR = np.sum((y_preds - np.mean(y))**2)

# Error Sum of Squares (SSE): unexplained variation in y (residuals)
SSE = np.sum((y - y_preds)**2)

# --- Model fit statistics ---

# Coefficient of determination (R^2): proportion of variation in y explained by the model
R2 = SSR / SST

# Adjusted R^2: adjusts for the number of predictors (p) relative to sample size (n)
R2_adj = 1 - (SSE / (n - p - 1)) / (SST / (n - 1))

# Overall F-statistic: tests H0 that all regression slopes = 0
# Numerator: mean regression sum of squares (SSR / p)
# Denominator: mean squared error (SSE / (n - p - 1))
F = (SSR / p) / (SSE / (n - p - 1))
F_crit = 3.59 # critical F value based on alpha = 0.05, p = 2, n-p-1 = 17

# Print multiple regression statistics in aligned columns
print(f"{'Statistic':<20} {'Value':>12}")
print(f"{'-'*32}")
print(f"{'R2':<20} {R2:12.4f}")
print(f"{'R2 Adjusted':<20} {R2_adj:12.4f}")
print(f"{'F-Statistic':<20} {F:12.4f}")
print(f"{'Significance (F-test)':<20} {str(F>F_crit):>11}")
print("-"*32)
print(f"SST: {SST}")
print(f"SSR: {SSR}")
print(f"SSE: {SSE}")
```

### Part C

```python
# --- Estimate of residual variance (sigma^2) ---
# Residual mean square: estimated variance of the errors
sigma = SSE / (n - p - 1)  # n = sample size, p = number of predictors

# --- Covariance matrix of regression coefficients ---
# Var(beta_hat) = sigma^2 * (X^T X)^(-1)
var_beta = sigma * np.linalg.inv(X.T @ X)  # diagonal entries give variance of each beta

# --- Standard errors of coefficients ---
# SE(beta_j) = sqrt(Var(beta_j))
SE_beta = np.sqrt(np.diagonal(var_beta))

# Extract standard errors for specific coefficients
SE_beta_6 = SE_beta[1]  # standard error for beta_6 (x6)
SE_beta_7 = SE_beta[2]  # standard error for beta_7 (x7)

# --- t-tests for individual coefficients ---
t_crit = 2.11  # critical t-value for two-sided test, df = n - p - 1 = 17

# Test statistics for each coefficient
t_beta_6 = beta_hat[1] / SE_beta_6  # t-statistic for beta_6
t_beta_7 = beta_hat[2] / SE_beta_7  # t-statistic for beta_7


print(f"t-statistic for x6: {t_beta_6:0.04f} | Significant: {t_beta_6>t_crit}")
print(f"t-statistic for x7: {t_beta_7:0.04f} | Significant: {t_beta_7>t_crit}")
print("-"*32)
print(f"Residual Variance: {sigma}")
print(f"SE Beta 6: {SE_beta_6}")
print(f"SE Beta 7: {SE_beta_7}")
```

### Part D

```python
# --- Function to calculate a two-sided confidence interval ---
def CI_calc(beta, t_crit, SE):
    """
    Computes the 95% confidence interval for a regression coefficient.
    
    Parameters:
    beta     : estimated coefficient (beta_hat)
    t_crit   : critical t-value for desired confidence level (e.g., 95%)
    SE       : standard error of the coefficient
    
    Returns:
    A tuple (lower_bound, upper_bound) representing the confidence interval
    """
    lower = beta - t_crit * SE  # lower bound of CI
    upper = beta + t_crit * SE  # upper bound of CI
    return lower, upper

# --- Calculate 95% confidence intervals for specific coefficients ---
CI_beta6 = CI_calc(beta_hat[1], t_crit, SE_beta_6)  # CI for beta_6 (x6)
CI_beta7 = CI_calc(beta_hat[2], t_crit, SE_beta_7)  # CI for beta_7 (x7)


print(f"Confidence Interval for Beta 6: {CI_beta6}")
print(f"Confidence Interval for Beta 7: {CI_beta7}")
```
