```python
## Import Library(s)
import pandas as pd
```

```python
## Question 1
## Read in survey_results_public.csv from 2022 Stack Overflow Survey https://insights.stackoverflow.com/survey 
df = pd.read_csv("survey_results_public.csv")
df.head()
```

```python
## Question 2
## Print out MainBranch types and their frequency
freq_dict = {}
for i in df["MainBranch"].unique():
    freq_dict[i] = 0

for i in df["MainBranch"]:
    if i in freq_dict.keys():
        freq_dict[i] += 1

freq_dict
```

```python
## Question 3
## Print out the length of your filtered list
max_key = ""
for key, value in freq_dict.items():
    if value == max(freq_dict.values()):
        max_key = key

df_most_responses = df[df["MainBranch"] == max_key]
df_most_responses.shape[0]
```

```python
## Question 4
## Print out the database types and their frequency
databases = []
for i in df_most_responses["DatabaseHaveWorkedWith"]:
    split_str = str(i).split(";")
    databases.extend(split_str)

database_count = dict((key,0) for key in set(databases))

for i in df_most_responses["DatabaseHaveWorkedWith"]:
    split_str = str(i).split(";")
    for i in split_str:
        database_count[i] += 1

database_count
```

```python
## Question 5 
## Create new column 'DBHWW_Count', sort and print out the count of the first record
df_most_responses = df_most_responses.reset_index()
df_most_responses["DBHWW_Count"] = 0
for idx, i in enumerate(df_most_responses["DatabaseHaveWorkedWith"]):
    split_str = str(i).split(";")
    df_most_responses.loc[df_most_responses.index==idx, "DBHWW_Count"] = len(split_str)

df_most_responses = df_most_responses.sort_values("DBHWW_Count", ascending=False)
df_most_responses.head(1)
```
