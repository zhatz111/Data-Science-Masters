```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Problem Statement
Consider a Baseball World Series (best of 7 game series) in which team A theoretically has a 0.75 chance of winning each game against team B. Games are considered independent. Simulate the probability that team A would win a World Series against team B by simulating 1000 World Series. You many use any software to conduct the simulation. This is easily derived from a binomial or negative binomial distribution, but I would like you to simulate it on a game x game basis. You may recognize that winning 4 in 4, 5, 6, or 7 games is equivalent to winning at least 4 in 7 games.

```python
def world_series():
    team_a_wins = 0
    team_b_wins = 0
    for i in range(7):
        game_winner = random.choices(
                population=["Team A", "Team B"],
                weights=[0.75, 0.25],
            )
        if game_winner[0] == "Team A":
            team_a_wins += 1
        else:
            team_b_wins += 1
        
        if team_a_wins == 4:
            return "Team A"
        if team_b_wins == 4:
            return "Team B"
        
```

```python
number_world_series = 1000

team_a_probability = []
team_a_world_series = 0
for i in range(number_world_series):
    winner = world_series()
    if winner == "Team A":
        team_a_world_series += 1
    team_a_probability.append(team_a_world_series/(i+1))

plt.plot(team_a_probability, "g")
plt.xlabel("Number of Simulations")
plt.ylabel("Probability")
plt.title(f"Team A to Win the World Series: {team_a_probability[-1]:.3f}")
```

```python
import numpy as np

def simulate_series(prob_A, num_series):
    wins_A = 0

    for _ in range(num_series):
        games_won_A = 0
        games_won_B = 0
        while games_won_A < 4 and games_won_B < 4:
            if np.random.rand() < prob_A:
                games_won_A += 1
            else:
                games_won_B += 1
        if games_won_A == 4:
            wins_A += 1

    return wins_A / num_series

# Parameters
prob_A = 0.75
num_series = 1000

# Simulate the World Series
prob_A_wins = simulate_series(prob_A, num_series)

print(f"Estimated probability that team A wins the World Series: {prob_A_wins:.4f}")
```
