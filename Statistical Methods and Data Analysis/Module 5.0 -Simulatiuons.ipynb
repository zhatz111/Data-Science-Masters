```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.random.binomial(10, 0.3, 500)
Y = np.random.binomial(5, 0.3, 500)
W = X + Y
W1 = np.random.binomial(15, 0.3, 500)
W.sort()
W1.sort()
```

```python
fig, ax = plt.subplots(1,2, figsize=(12,5))
ax[0].scatter(W, W1)
ax[0].set_title("Scatter Plot W vs W1")
ax[0].set_xlabel("W")
ax[0].set_ylabel("W1")

ax[1].hist((W,W1), label=("W","W1"))
ax[1].set_title("Histogram Plot W vs W1")
ax[1].set_xlabel("Bins")
ax[1].set_ylabel("Frequency")
ax[1].legend()
```

```python
n_samples = 2000
lambda_ = 1
X = np.random.exponential(1/lambda_, n_samples)
Y = np.random.exponential(1/lambda_, n_samples)
W = Y/X
r = np.random.uniform(0, 1, n_samples)
Wdev = (1/(1-r))-1
W.sort()
Wdev.sort()

bot_95 = int(round(W.shape[0]*0.95,0))
W = W[:bot_95]
Wdev = Wdev[:bot_95]
```

```python
plt.scatter(W, Wdev)
plt.xlabel("W")
plt.ylabel("Wdev")
plt.title("Scatter plot of W vs Wdev")
# plt.xlim(0,10)
# plt.ylim(0,3)
```
