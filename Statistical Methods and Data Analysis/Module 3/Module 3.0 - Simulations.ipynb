```python
import numpy as np
import matplotlib.pyplot as plt
```

### 1. Using the rbinom function, generate 1000 random deviates from a B(20,.4) distribution, compute the sample mean and compare with the theoretical expectation.

```python
random_draws_binom = np.random.binomial(20,0.4, 1000)
theoretical_binom = 20*0.4
print(f"Mean of Random Draws: {np.mean(random_draws_binom)}")
print(f"Theoretical Expectation: {theoretical_binom}")
```

### 2. Using the rhyper function, generate 1000 random deviates from a hypergeometric distribution with 4 white balls, 8 black balls, and 3 balls drawn without replacement. Compute the sample mean and compare with the theoretical expectation.

```python
random_draws_hyper = np.random.hypergeometric(4,8,3,1000)
theoretical_hyper = 3 * (4/(4+8))
print(f"Mean of Random Draws: {np.mean(random_draws_hyper)}")
print(f"Theoretical Expectation: {theoretical_hyper}")
```

### 3. Using the rexp function, generate 1000 random deviates from an exponential with lambda=3. Compute the sample mean and compare with the theoretical expectation, which is 1/lambda.

```python
random_draws_exp = np.random.exponential(1/3,1000)
theoretical_exp = 1/3
print(f"Mean of Random Draws: {np.mean(random_draws_exp)}")
print(f"Theoretical Expectation: {theoretical_exp}")
```

```python
# Define the x values
x = np.linspace(0.01, 10, 1000)

# Define the exponential density functions
y1 = 4 * np.exp(-4 * x)
y2 = 2 * np.exp(-2 * x)
y3 = np.exp(-x)
y4 = 0.5 * np.exp(-0.5 * x)

# Plot the first exponential density function
plt.plot(x, y1, label='4*exp(-4*x)', color='black')

# Plot the second exponential density function
plt.plot(x, y2, label='2*exp(-2*x)', color='red')

# Plot the third exponential density function
plt.plot(x, y3, label='1*exp(-x)', color='blue')

# Plot the fourth exponential density function
plt.plot(x, y4, label='0.5*exp(-0.5*x)', color='green')

# Add labels and legend
plt.xlabel('x')
plt.ylabel('Density')
plt.legend()
```
