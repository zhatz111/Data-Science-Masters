```python
import random
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# List out rate data from 9.2.20
fixed_30_year = [3.525,3.625,3.383,3.625,3.661,3.791,3.941,3.781,3.660,3.733]
arm = [2.923,3.385,3.154,3.363,3.226,3.283,3.427,3.437,3.746,3.438]
combined = fixed_30_year + arm
initial_mean = np.mean(fixed_30_year) - np.mean(arm)

# Simulate various permuations of combined data
perms = 1000
alpha = 0.05
mean_diff_permute = []
for i in range(perms):
    first_sample = random.sample(combined, 10)
    second_sample = list(set(combined) - set(first_sample))
    mean_diff = np.mean(first_sample) - np.mean(second_sample)
    mean_diff_permute.append(mean_diff)

# Calculate the p-value and graph the mean data
greater_than_initial = len([i for i in mean_diff_permute if i > initial_mean])
p_value = greater_than_initial/perms
sns.histplot(mean_diff_permute)
plt.xlabel("Mean Difference")
plt.title(f"p-value = {p_value}")
```
