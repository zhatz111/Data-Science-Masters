```python
import numpy as np
import matplotlib.pyplot as plt

n_samples = 1000
n_rand_vars = 3
p_mean = 2
exp_rate = 0.25 
poisson_list = [np.sum(np.random.poisson(p_mean,n_rand_vars)) for _ in range(n_samples)]
exp_list = [np.sum(np.random.exponential(exp_rate,n_rand_vars)) for _ in range(n_samples)]
poisson_list.sort()
exp_list.sort()

fig, ax = plt.subplots(1,2, figsize=(12,5))
ax[0].plot(poisson_list[51:950], "o", markersize=2)
ax[0].set_title("Sum of 3 Poisson Random Variables")
ax[0].set_xlabel("Count")
ax[0].set_ylabel("Random Deviates")

ax[1].plot(exp_list[51:950], "o", markersize=2)
ax[1].set_title("Sum of 3 Exponential Random Variables")
ax[1].set_xlabel("Count")
ax[1].set_ylabel("Random Deviates")
```
