```python
import numpy as np
from matplotlib import pyplot as plt 

n_sims = 10000
below_5 = 0
below_5_list = []
for i in range(n_sims):
    rand_nums = np.random.uniform(0,1,8)
    rand_nums.sort()
    # rand_nums = rand_nums[::-1]
    if rand_nums[5] < 0.5:
        below_5 += 1
    if i != 0:
        below_5_list.append(below_5/i)

plt.plot(below_5_list)
plt.title(f"Probability is {below_5/10000}")
plt.xlabel("Number of Simulations")
plt.ylabel("Probability")
```

```python
step_10 = 0
step_0 = 0

for _ in range(10000):
    initial = 5
    for step in range(10):
        next_step = np.random.choice([-1,1], p=[0.4,0.6])
        initial += next_step
    if initial >= 10:
        step_10 += 1
    if initial <= 0:
        step_0 += 1

print(f"Lower: {step_0/10000}, Upper: {step_10/10000}")
```
