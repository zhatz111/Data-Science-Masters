```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mc_data_1 = pd.read_csv("MCdata1.txt", header=None, delimiter=" ")
mc_data_2 = pd.read_csv("MCdata2.txt", header=None, delimiter=" ")
```

```python
mc_chain_1 = []
mc_chain_2 = []
for row in mc_data_1.iterrows():
    mc_chain_1.extend(row[1].values)

for row in mc_data_2.iterrows():
    mc_chain_2.extend(row[1].values)
```

### Markov Chain Monte Carlo Plot

```python
plt.figure(figsize=(15,3))
plt.plot(mc_chain_1)
```

```python
plt.figure(figsize=(15,3))
plt.plot(mc_chain_2)
```

### Cumulative Sum (Cusum diagnostic)

```python
cusum = 0
chain_1_avg = np.mean(mc_chain_1)
cusum_chain_1 = []
for i in mc_chain_1:
    cusum = cusum + (i - chain_1_avg)
    cusum_chain_1.append(cusum)

plt.figure(figsize=(15,3))
plt.plot(cusum_chain_1)
```

```python
cusum = 0
chain_2_avg = np.mean(mc_chain_2)
cusum_chain_2 = []
for i in mc_chain_2:
    cusum = cusum + (i - chain_2_avg)
    cusum_chain_2.append(cusum)

plt.figure(figsize=(15,3))
plt.plot(cusum_chain_2)
plt.axhline(0)
```
