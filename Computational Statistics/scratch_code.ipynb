```python
import matplotlib.pyplot as plt
import pandas as pd
from pandas.plotting import parallel_coordinates

# Create the dataset
data = {
    "Obs": ["1", "2", "3", "4"],
    "x1": [6, 1, 4, 2],
    "x2": [5, 2, 4, 2],
    "x3": [4, 3, 6, 2],
    "x4": [3, 4, 4, 3]
}

df = pd.DataFrame(data)

# Plot parallel coordinates
plt.figure(figsize=(10, 6))
parallel_coordinates(df, class_column="Obs", colormap=plt.cm.Set1)
plt.title("Parallel Coordinate Plot of Observations")
plt.xlabel("Variables")
plt.ylabel("Values")
plt.grid(True)
plt.show()
```

```python
final_price_ = 0
for i in range(1, 21):
    price_ = 5/(1.08)**i
    final_price_ += price_
complete_price = final_price_ + 100/(1.08)**20
complete_price
```

```python
import numpy as np
def func_(r):
    final_price = 0
    for i in range(1, 11):
        price = 5/(1 + r)**i
        final_price += price
    return final_price + 110/(1 + r)**10 - complete_price

x_vals = np.linspace(.10, 0.11, 1000000)
vals = []
for i in x_vals:
    vals.append(func_(i))
```

```python
import matplotlib.pyplot as plt

plt.plot(x_vals, vals)
plt.axhline(0, linestyle="--")
```

```python
guess = x_vals[np.where(np.array(vals)<0)[0][0]]
func_(guess)
```
