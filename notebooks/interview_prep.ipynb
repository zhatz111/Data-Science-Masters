## Linear Regression Math

```python
import pandas as pd
import numpy as np

data = pd.read_csv(r"C:\Users\zhatz\Downloads\weatherHistory.csv")
reduced = data.loc[:, "Temperature (C)":"Pressure (millibars)"]
X = reduced.loc[:, reduced.columns!="Apparent Temperature (C)"].values
y = reduced.loc[:, "Apparent Temperature (C)"].values

coeff = np.linalg.pinv(X.T @ X) @ X.T @ y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=False)
model.fit(X, y)
beta = model.coef_
beta
```
