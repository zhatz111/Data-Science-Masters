```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("FastFoodNutritionMenuV3.csv")
df = df.drop(["Calories from Fat", "Weight Watchers Pnts"], axis=1)
df.head()
```

```python
for col in df.columns[2:]:
    temp_list = []
    for val in df[col]:
        try:
            temp_list.append(float(val))
        except ValueError:
            temp_list.append(np.nan)
    df[col] = temp_list
```

```python
df = df.dropna()
df.head()
```
